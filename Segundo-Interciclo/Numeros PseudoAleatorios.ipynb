```python
#Importacion de librerias a utilizar
import psutil
import cpuinfo

tem= 1077
OxC= 27
nprox = float(cpuinfo.get_cpu_info()['count'])
vel = float(cpuinfo.get_cpu_info()['hz_advertised_friendly'].split(" ")[0])
flops = nprox*vel*OxC
mflops = int(flops*1024)
uso_cpu = psutil.cpu_percent()
ram = psutil.virtual_memory().total
uso_ram = psutil.virtual_memory().used
valores = [tem, mflops, uso_cpu, ram, uso_ram, 185925, 21041, 8265, 261119,20127]

def obtenerN (digitos):
    valor1= 0
    valor2= 0
    
    if digitos%2 !=0:
        valor1 = int(digitos/2)
        valor2 = int(digitos/2)+1
    else:
        valor1 = int(digitos/2)
        valor2 = valor1
    return valor1,valor2

def calcular(iters, val, digitos):
    x0_semilla = int(val)
    aum = obtenerN(digitos)
    print("Iteracion: ", "Xn", "Xn*Xn", "Longitud","Ui","Rn")
    for i in range(iters):
        xn2= x0_semilla**2
        long = len(str(xn2))
        ui = str(xn2)[int(long/2)-aum[0]:int(long/2)+aum[1]]
        rn = int(ui)/10**digitos
        print(i, " ", x0_semilla," ",xn2, " ", long, " ",ui, " ", rn)
        x0_semilla=int(ui)
    print(" ")

    
```

```python
iteraciones = 16
digitos = 4
for i in valores:
    print("i: ", i)
    calcular(iteraciones, i,digitos)
    print(" ")
```
