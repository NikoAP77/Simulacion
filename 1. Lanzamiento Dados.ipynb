## Simular el evento de lanzar dos dados

```python
from collections import Counter
import json
import random
import matplotlib.pyplot as plt

suma=[]

def lanzamientoDados(numLanzamientos):
    for i in range(numLanzamientos):
        dado1 = random.randint(1,6)
        dado2 = random.randint(1,6)
        suma.append(dado1+dado2)
    numSuma=Counter(suma)
    

    plt.hist(suma,bins=numLanzamientos, width=0.5, alpha=0.7)
    plt.grid(axis='y', alpha=0.75)
    plt.xlabel('Valor',fontsize=15)
    plt.ylabel('Frequencia',fontsize=15)
    plt.xticks(range(0, max(numSuma.keys())+1))
    plt.show()
    
    for i in sorted(numSuma.items()):
        print("Total de la suma: ", i[0], "Total de Probabilidad: ", round((i[1]/numLanzamientos)*100,2), "%")
    print()
    print("Suma: ", numSuma)
  
     
```

```python
lanzamientoDados(100)
```

```python
lanzamientoDados(1000)
```

```python
lanzamientoDados(10000)
```
